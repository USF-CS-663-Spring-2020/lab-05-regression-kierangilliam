```python
from typing import List
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import numpy as np

DEBUG = False

def debug(item):
    if DEBUG:
        print(item)
```

```python
def linear_regression(
        filename: str,        
        predictors: List[str], 
        target: str,
):

    df = pd.read_csv(filename)
    
    for predictor in predictors:
        df.plot(kind='scatter', x=predictor, y=target)
        plt.show()
        resid = sns.residplot(target, predictor, data=df)
        resid.set(xlabel='Limit', ylabel='Residual')
        plt.show()

    debug(df.head())
    
    for predictor in predictors:
        reg = LinearRegression()
        reg.fit(df[[predictor]], df[[target]])
        print(f"""
        {predictor} against {target} 
            Slope: {reg.coef_[0][0]}
            Intercept: {reg.intercept_}
        """)
```

```python
linear_regression('toluca.csv', ['workHours'], 'lotSize')

# Work hours is a pretty bad predictor
```

```python
predictors = ['Income', 'Rating', 'Cards', 'Age', 'Education']
linear_regression('Credit.csv', predictors, 'Limit')

# Rating and income are good predictors, rest are bad 
```
